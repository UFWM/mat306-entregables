#### Dataframe con la cual se trabajará

Ahora, queremos sacar la informacion del desglose de conexiones a sitios web durante el 2020, para cada distrito en nuestra dataframe. Para esto, transformemos los distintos IDs de distrito en el formato del nombre de los archivos en los que se encuentran. Luego, podemos para cada uno de estos archivos generar su respectiva dataframe, para luego concatenarlas (manteniendo registro de donde vienen los datos) y asi solo trabajar con una dataframe. Aseguremonos de considerar solamente los datos que tienen "lp_id" no vacio, pues no nos sirven para poder identificar de que sitio web provienen. Además, una vez juntados todos estos datos en un dataframe, podemos olvidarnos del "lp_id" y así solo hablar del nombre de las aplicaciones.

```python
districts=df_districts_filter['district_id'].unique().tolist()
districts_path=[str(x)+".csv" for x in districts]
districts_list=[]
i=0

for district_path in districts_path:
    archive_path = os.path.join('engagement_data', district_path)
    df_temp=pd.read_csv(archive_path)
    if districts[i] in districts:
        df_temp['state']=df_districts_filter[df_districts_filter['district_id']==districts[i]]['state'].squeeze()
        df_temp['locale']=df_districts_filter[df_districts_filter['district_id']==districts[i]]['locale'].squeeze()
    else:
        df_temp['state']=np.NAN
        df_temp['locale']=np.NAN
    districts_list.append(df_temp)
    i=i+1

df_1=pd.concat(districts_list).reset_index(drop=True)
df_1
```

```python
df_2=pd.DataFrame(data=zip(*[df_products.index.tolist(),df_products['Product Name'].tolist()]),columns=['lp_id','product_name'])
df=pd.merge(df_1,df_2,how='inner',on='lp_id').drop(columns='lp_id')
df
```

```python
df.info()
```

Estudiemos un poco los dataset entregados previo a trabajar. Veamos primero el conjunto de entrenamiento.

```python
df_train
```

Podemos ver que hay algunas columnas que le faltan datos y hay columnas no numéricas en el dataset. Arreglemos esto para poder entrenar bien el modelo a futuro. Para esto, descartemos ciertas columnas que presentan bastantes faltas de datos o parecen ser irrelevantes para la supervivencia. Por ejemplo, la columna "Cabin" tiene muchos datos nulos, "Name" no debería influenciar porque el nombre de las personas no debería afectar la supervivencia, y similarmente con "PassengerID" y "Ticket". El resto de las columnas se transformará en tipo "int64" para facilitar el tratamiento de datos mediante "pd.get_dummies". Ademas, botaremos todas las columnas que presentan un valor nulo, para no tener problemas para el entrenamiento.

```python
df_train_filter=df_train.drop(columns=['PassengerId','Cabin','Name','Ticket'],axis=1).dropna()
df_train_final=pd.get_dummies(df_train_filter, columns=['Sex','Embarked'], prefix='', prefix_sep='', dtype=int)

df_train_final
```

Visualizemos la correlación entre las variables en el conjunto de entrenamiento.

```python
sns.heatmap(df_train_final.corr(), annot=True, cmap='PiYG', fmt='.2f')
```

No podemos ver patrones evidentes, salvo que hay una relación bastante mayor entre la supervivencia con el sexo de la persona,y que las columnas dadas por "get_dummies" tienen alta correlación de manera esperada. Ahora separemos en conjuntos para entrenar los datos con algunos modelos de ML.

```python
X = df_train_final.drop('Survived', axis=1)
y = df_train_final['Survived']

X_train, X_eval, y_train, y_eval = train_test_split(X, y, test_size=0.25, random_state=1)
```

Ocuparemos tres tipos de modelos distintos y compararemos sus desempeños. De estos tres modelos se intentará hacer una optimización de hiperparámetros para encontrar el mejor de su clase, para compararla con el resto.

**(1) Regresión Lineal**

```python
# Definir el modelo base
model = LogisticRegression()

# Rango de parámetros para Grid Search
param_grid = {
    'C': [0.001, 0.01, 0.1, 1, 10, 100],
    'max_iter': [100, 200, 500, 1000]
}

# Configuración y aplicación de Grid Search
gs = GridSearchCV(
    estimator=model, 
    param_grid=param_grid, 
    scoring='accuracy',    # Métrica de evaluación
    cv=5,                  # Número de particiones de validación cruzada
    n_jobs=-1              # Usar todos los núcleos disponibles para acelerar el proceso
)

# Entrenamiento de Grid Search
gs.fit(X_train, y_train)

# Resultados del mejor modelo
print("Mejor precisión en validación cruzada:", gs.best_score_)
print("Mejores parámetros:", gs.best_params_)

# Entrenar y evaluar el mejor modelo en el conjunto de evaluación
mejor_modelo_lr = gs.best_estimator_
mejor_modelo_lr.fit(X_train, y_train)
print("Precisión en el conjunto de evaluación: {:.3f}".format(mejor_modelo_lr.score(X_eval, y_eval)))
```

**(2) Support Vector Machine**

```python
# Definir el modelo base
model = SVC()

# Rango de parámetros para Grid Search
param_grid = {
    'C': [0.001, 0.01, 0.1, 1, 10, 100],
    'degree': [1, 2, 3]
}

# Configuración y aplicación de Grid Search
gs = GridSearchCV(
    estimator=model, 
    param_grid=param_grid, 
    scoring='accuracy',    # Métrica de evaluación
    cv=5,                  # Número de particiones de validación cruzada
    n_jobs=-1              # Usar todos los núcleos disponibles para acelerar el proceso
)

# Entrenamiento de Grid Search
gs.fit(X_train, y_train)

# Resultados del mejor modelo
print("Mejor precisión en validación cruzada:", gs.best_score_)
print("Mejores parámetros:", gs.best_params_)

# Entrenar y evaluar el mejor modelo en el conjunto de evaluación
mejor_modelo_svc = gs.best_estimator_
mejor_modelo_svc.fit(X_train, y_train)
print("Precisión en el conjunto de evaluación: {:.3f}".format(mejor_modelo_svc.score(X_eval, y_eval)))
```

**(3) Random Forest**

```python
# Definir el modelo base
model = RandomForestClassifier(random_state=1)

# Rango de parámetros para Grid Search
param_grid = {
    'n_estimators': [10, 20, 50, 100, 150, 200],
    'max_depth': [5, 10, 15, 20, 25],
}

# Configuración y aplicación de Grid Search
gs = GridSearchCV(
    estimator=model, 
    param_grid=param_grid, 
    scoring='accuracy',    # Métrica de evaluación
    cv=5,                  # Número de particiones de validación cruzada
    n_jobs=-1              # Usar todos los núcleos disponibles para acelerar el proceso
)

# Entrenamiento de Grid Search
gs.fit(X_train, y_train)

# Resultados del mejor modelo
print("Mejor precisión en validación cruzada:", gs.best_score_)
print("Mejores parámetros:", gs.best_params_)

# Entrenar y evaluar el mejor modelo en el conjunto de evaluación
mejor_modelo_rf = gs.best_estimator_
mejor_modelo_rf.fit(X_train, y_train)
print("Precisión en el conjunto de evaluación: {:.3f}".format(mejor_modelo_rf.score(X_eval, y_eval)))
```

Podemos ver que el modelo de Random Forest resulto ser el mejor en clasificar de los 3, ligeramente mejor que la regresión lineal.
